```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
rows = 100000
```

```python
campaign_types = [
    'Credit Card',
    'Personal Loan',
    'Home Loan',
    'Savings Account',
    'Insurance'
]
```

```python
channels = [
    'Email',
    'SMS',
    'Mobile App',
    'Web'
]
```

```python
regions = [
    'North',
    'South',
    'East',
    'West'
]
```

```python
data = pd.DataFrame({

    'Customer_ID': np.arange(1, rows+1),

    'Age': np.random.randint(18,65,rows),

    'Gender': np.random.choice(
        ['Male','Female'],
        rows
    ),

    'Income': np.random.randint(
        25000,
        150000,
        rows
    ),

    'Region': np.random.choice(
        regions,
        rows
    ),

    'Tenure': np.random.randint(
        1,
        15,
        rows
    ),

    'Product_Count': np.random.randint(
        1,
        6,
        rows
    ),

    'Digital_User': np.random.choice(
        ['Yes','No'],
        rows,
        p=[0.75,0.25]
    ),

    'Campaign_ID': np.random.randint(
        1000,
        1050,
        rows
    ),

    'Campaign_Type': np.random.choice(
        campaign_types,
        rows
    ),

    'Channel': np.random.choice(
        channels,
        rows
    ),

    'Budget': np.random.randint(
        50000,
        500000,
        rows
    )
})

data['Opened'] = np.random.choice(
    [0,1],
    rows,
    p=[0.4,0.6]
)

data['Clicked'] = np.where(
    data['Opened']==1,
    np.random.choice(
        [0,1],
        rows,
        p=[0.5,0.5]
    ),
    0
)

data['Converted'] = np.where(
    data['Clicked']==1,
    np.random.choice(
        [0,1],
        rows,
        p=[0.7,0.3]
    ),
    0
)

data['Revenue'] = np.where(
    data['Converted']==1,
    np.random.randint(
        1000,
        15000,
        rows
    ),
    0
)

data.to_csv(
    'customer_campaign_data.csv',
    index=False
)

print("Dataset Created Successfully")
print(data.head())
```

```python
import pandas as pd

df = pd.read_csv("customer_campaign_data.csv")

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nMissing Values")
print(df.isnull().sum())

print("\nConversion Rate")
print(df['Converted'].mean()*100)

print("\nOpen Rate")
print(df['Opened'].mean()*100)

print("\nCTR")
print(df['Clicked'].sum()/df['Opened'].sum()*100)
```

```python
df = pd.read_csv("customer_campaign_data.csv")

total_customers = df['Customer_ID'].nunique()
total_revenue = df['Revenue'].sum()

open_rate = round(
    df['Opened'].mean()*100,2
)

ctr = round(
    (df['Clicked'].sum()/df['Opened'].sum())*100,2
)

conversion_rate = round(
    df['Converted'].mean()*100,2
)

print("Total Customers:", total_customers)
print("Revenue:", total_revenue)
print("Open Rate:", open_rate)
print("CTR:", ctr)
print("Conversion Rate:", conversion_rate)
```
